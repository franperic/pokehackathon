```python
def create_master_table(
    all_pokemons: pd.DataFrame, available_pokemons: pd.DataFrame, weakness_pokemons: pd.DataFrame
) -> pd.DataFrame:
    """Combines all data to create a master table.

        Args:
            all_pokemons: Preprocessed data for All Pokemons.
            available_pokemons: Preprocessed data for Available Pokemons.
            weakness_pokemons: Preprocessed data for Weakness Pokemons.
        Returns:
            Master table.

    """
#     available_pokemons_with_ID_and_types = all_pokemons.merge(available_pokemons, left_on="Name", right_on="Name_1")
    available_pokemons_with_ID_and_types = available_pokemons.merge(all_pokemons, left_on="Name_1", right_on="Name")
    with_weakness_1 = available_pokemons_with_ID_and_types.merge(weakness_pokemons, left_on='Type_1', right_on='Types')

#     with_companies = rated_shuttles.merge(
#         companies, left_on="company_id", right_on="id"
#     )

#     master_table = with_companies.drop(["shuttle_id", "company_id"], axis=1)
#     master_table = master_table.dropna()
    return with_weakness_1#master_table
```
